```python
def preprocess(df):
    occupation_type_map = {'Accountants': 0,
                         'Cleaning staff': 1,
                         'Cooking staff': 2,
                         'Core staff': 3,
                         'Drivers': 4,
                         'HR staff': 5,
                         'High skill tech staff': 6,
                         'IT staff': 7,
                         'Laborers': 8,
                         'Low-skill Laborers': 9,
                         'Managers': 10,
                         'Medicine staff': 11,
                         'Private service staff': 12,
                         'Realty agents': 13,
                         'Sales staff': 14,
                         'Secretaries': 15,
                         'Security staff': 16,
                         'Unknown': 17,
                         'Waiters/barmen staff': 18}
    
    gender_map = {'F' : 0, 'M' : 1}
    owns_car_map = {'Y' : 1, 'N' : 0}
    owns_house_map = {'Y' : 1, 'N' : 0}
    
    df['occupation_type'] = df['occupation_type'].map(occupation_type_map)
    df['gender'] = df['gender'].map(gender_map)
    df['owns_car'] = df['owns_car'].map(owns_car_map)
    df['owns_house'] = df['owns_house'].map(owns_house_map)
    
    # return df
```

```python
preprocess(df)
```

# train rf

```python
from sklearn.ensemble import RandomForestClassifier
import numpy as np
from sklearn.metrics import f1_score
```

```python
train_df.info()
```

```python
X = train_df.fillna(0).drop(['credit_card_default', 'customer_id', 'name'], axis=1).values.astype('float32')

y = train_df['credit_card_default'].values.astype('float32')
```

```python
clf = RandomForestClassifier(max_depth=10, n_estimators=50, random_state=0)
```

```python
clf.fit(X, y)
```

```python
X_val = val_df.fillna(0).drop(['credit_card_default', 'customer_id', 'name'], axis=1).values.astype('float32')
y_val = val_df['credit_card_default'].values.astype('float32')
```

```python
pred = clf.predict(X_val)
```

```python
f1_score(y_val, pred, average='macro')
```
